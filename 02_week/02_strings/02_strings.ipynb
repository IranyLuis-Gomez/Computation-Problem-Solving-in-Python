```python
# Access string elements
name = "Konstantin"
print(name[0])
print(name[1])

# Use negative indices
print(name[-1]) # prints last char of the str

print(name[-3]) 
```

```python
# string slicing 
#specify the start and the end of the slice 
language = "Pytnon"
language = language.replace("n", "h", 1)
print(language)
print(language[0:3])

# what is the output 
print(language[2:len(language)])
print(language[2:])
print(language[:5])
```

```python
# you can use negative indices 
language = "Python"

print(language[-3:])

# specify the step
print(language[::2])
print(language[::-2])

#reverse a string 
print(language[::-1])
```

```python
text = "123456789"
# print second half
print(text[len(text) // 2:])

# print first 3 char plus last 2 char
print(text[:3] + text[-2:])
```

```python
# split
makes = "toyota, bmw, audi, volkswagon, nissan"

car_makes = makes.split(",") #car_makes is a list of strings

# iterate through the list
for make in car_makes:
    make = make.lstrip()
    if make == "bmw" or make == "audi":
        print(make.upper())
    else:
        print(make.title())
```
